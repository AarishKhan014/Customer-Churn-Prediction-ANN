## Prediction

```python
import pandas as pd
import pickle
import numpy as np
```

```python
# Load trained model

from tensorflow.keras.models import load_model

model = load_model('model.h5')

# Load scaler pickle, encoding files

with open('onehot_encoder_gender.pkl', 'rb') as file:
    onehot_encoder_gender = pickle.load(file)

with open('onehot_encoder_geography.pkl', 'rb') as file:
    onehot_encoder_geography = pickle.load(file)

with open('standard_scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender' : 'Male',
    'Age' : 40,
    'Tenure' : 3,
    'Balance' : 60000,
    'NumOfProducts' : 2,
    'HasCrCard' : 1,
    'IsActiveMember' : 1,
    'EstimatedSalary' : 50000
}
```

```python
# Transforming And Concatenating Inputs

gen_encoded = onehot_encoder_gender.transform([[input_data['Gender']]]).toarray()
gen_encoded_df = pd.DataFrame(gen_encoded, columns=onehot_encoder_gender.get_feature_names_out())

geo_encoded = onehot_encoder_geography.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geography.get_feature_names_out())

input_df = pd.concat([pd.DataFrame(input_data, index=[0]), geo_encoded_df, gen_encoded_df], axis=1).drop(columns=['Geography', 'Gender'])
```

```python
# Scaling Input

input_scaled = scaler.transform(input_df)
input_scaled
```

```python
# Predict Churn

prediction = model.predict(input_scaled)
prediction
```

```python
pred_proba = prediction[0][0]
pred_proba
```

```python
if pred_proba > 0.5:
    print ('The Customer Is Likely To To Churn')
else:
    print ('The Customer Is Not Likely To Churn')
```
